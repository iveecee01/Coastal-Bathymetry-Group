```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import rasterio as rio
import earthpy.plot as ep
```

```python
filepath = "../RawData/S1AA_20181008T230552_20181125T230551_VVP048_INT80_G_weF_61C9/S1AA_20181008T230552_20181125T230551_VVP048_INT80_G_weF_61C9_vert_disp.tif"
with rio.open(filepath) as vert_disp:
    vert = vert_disp.read(1)
```

```python
type(vert)
```

```python
ep.plot_bands(vert,
              title="Vertical Displacement in Outer Banks \n before and after Hurricane Michael (2018)",
              cmap="OrRd")

plt.show()
```
